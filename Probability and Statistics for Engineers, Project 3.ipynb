```python
import seaborn as sns
import matplotlib.pyplot as plt

#sample size = 10

data = [23.6294896, 24.05693475, 20.0617284, 21.46193662, 25, 22.75830678, 22.1526465, 17.96875, 22.265625, 19.70553242]


#histogram with normalized probabilities
sns.histplot(data, kde=False, stat='probability')
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Probability Distribution (BMI n=10)')
plt.show()
```
