```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = [23.6294896, 24.05693475, 20.0617284, 21.46193662, 25, 22.75830678, 22.1526465, 17.96875, 22.265625, 19.70553242]


# lower and upper bounds of the confidence interval
lower_bound = 20.6534077816
upper_bound = 23.1587822384

# density plot is created based on the data
sns.kdeplot(data)

# filling-in the area between the x-axis and the confidence interval bounds
# sns.kdeplot(data).get_lines()[0].get_data()[1] extracts the y-coordinates
plt.fill_betweenx([0, max(sns.kdeplot(data).get_lines()[0].get_data()[1])],
                  lower_bound, upper_bound, alpha=0.4, color='red')

plt.text(lower_bound, 0, f'{lower_bound:.3f}', ha='right', va='bottom', color='red')
plt.text(upper_bound, 0, f'{upper_bound:.3f}', ha='left', va='bottom', color='red')



# Add labels and title to the plot
plt.xlabel('Data')
plt.ylabel('Density')
plt.title('90% Confidence Interval for the Mean of BMI (n=10)')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Set the degrees of freedom
df = 9

# Generate x values for the t-distribution
x = np.linspace(-4, 4, 100)

# Calculate the probability density function (PDF) of the t-distribution
pdf = t.pdf(x, df)

t_critical_left = -2.262156
t_critical_right = 2.262156

# plotting the t-distribution
plt.plot(x, pdf, label='t-distribution (df=10)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('T-Distribution')

# Mark the critical regions
plt.axvline(t_critical_left, color='red', linestyle='--', label='Critical Region')
plt.axvline(t_critical_right, color='red', linestyle='--')

# Shade the critical regions
plt.fill_between(x, 0, pdf, where=(x < t_critical_left) | (x > t_critical_right), color='red', alpha=0.9)



# Specify rejection and non-rejection areas
plt.text(t_critical_left, 0, f'{t_critical_left:.2f}', ha='right', va='bottom', color='white')
plt.text(t_critical_right, 0, f'{t_critical_right:.2f}', ha='left', va='bottom', color='white')

plt.annotate('Rejection Region (0.025)', xy=(2.5, 0.02), xytext=(2.3, 0.15),
             arrowprops=dict(facecolor='green', arrowstyle='->'))
plt.annotate('Rejection Region (0.025)', xy=(-2.5, 0.02), xytext=(-4.25, 0.15),
             arrowprops=dict(facecolor='green', arrowstyle='->'))


plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm


# Generate x values for the t-distribution
x = np.linspace(-4, 4, 100)

# Calculate the probability density function (PDF) of the t-distribution
pdf = norm.pdf(x, 0, 1)

t_critical_left = -1.96
t_critical_right = 1.96

# plotting the t-distribution
plt.plot(x, pdf, label='Z-distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')

# Mark the critical regions
plt.axvline(t_critical_left, color='red', linestyle='--', label='Critical Region')
plt.axvline(t_critical_right, color='red', linestyle='--')

# Shade the critical regions
plt.fill_between(x, 0, pdf, where=(x < t_critical_left) | (x > t_critical_right), color='red', alpha=0.9)



# Specify rejection and non-rejection areas
plt.text(t_critical_left, 0, f'{t_critical_left:.2f}', ha='right', va='bottom', color='white')
plt.text(t_critical_right, 0, f'{t_critical_right:.2f}', ha='left', va='bottom', color='white')

plt.annotate('Rejection Region (0.025)', xy=(2.5, 0.02), xytext=(2.15, 0.15),
             arrowprops=dict(facecolor='green', arrowstyle='->'))
plt.annotate('Rejection Region (0.025)', xy=(-2.5, 0.02), xytext=(-4.25, 0.15),
             arrowprops=dict(facecolor='green', arrowstyle='->'))

plt.legend()
plt.grid(True)
plt.show()
```
